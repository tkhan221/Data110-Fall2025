```python
import matplotlib.pyplot as plt
import seaborn as sns
pg = sns.load_dataset("penguins")
species_mean = pg.groupby('species')['body_mass_g'].mean()
plt.show()
plt.bar(species_max.index, species_max.values, color='teal')
plt.xlabel('Species')
plt.ylabel('Maximum Body Mass (g)')
plt.title('Maximum Body Mass by Penguin Species')
plt.show()
```

```python
sns.barplot(x='species')
```

```python
species_sex_grouped=pg.groupby(['species','sex'])['body_mass_g'].mean()
print(species_sex_grouped)
```

```python
species_sex_grouped=pg.groupby(['species','sex'])['body_mass_g'].mean().unstack()
print(species_sex_grouped) #this is multiseries but pandas multiseries
```

```python
species_sex_grouped.plot(kind='bar')
```
