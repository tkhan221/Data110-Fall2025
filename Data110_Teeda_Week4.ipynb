```python

df_pct = df.iloc[2]
```

```python
crime_number_columns = [
    'murder_number',
    'rape_number',
    'robbery_number',
    'aggravatedassault_number',
    'violentcrime_number',
    'burglary_number',
    'larceny_number',
    'motorvehicletheft_number',
    'propertycrime_number'
]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


values = df_pct[crime_number_columns].values
labels = [col.replace('_number','').capitalize() for col in crime_number_columns]
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=values, y=labels, palette='coolwarm')
plt.xlabel('% Change (2022  2023)')
plt.title('Percentage Change in Crimes (2022  2023)')
plt.axvline(0, color='black', linewidth=0.8)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example data
data = {
    'Company': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Product': ['Product 1', 'Product 2', 'Product 3'] * 3,
    'Sales': [130, 220, 200, 110, 220, 190, 130, 170, 160]
}

df_fake = pd.DataFrame(data)

plt.figure(figsize=(10,6))
sns.barplot(x='Company', y='Sales', hue='Product', data=df_fake, palette='Set2')
plt.title('Sales by Company and Product')
plt.ylabel('Sales')
plt.xlabel('Company')
plt.legend(title='Product')
plt.show()

```
